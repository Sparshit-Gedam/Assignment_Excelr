```python
X = df.drop(columns = 'Type')
Y = df['Type']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.95)  # 95% variance
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
X_pca[:5], explained_variance, X_pca.shape
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=Y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Wine Dataset (First 2 Components)')
plt.colorbar(scatter, label='Wine Type')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Define number of clusters based on known wine types (3 types in this dataset)
n_clusters = 3

# Apply K-means clustering on the original data (X_scaled)
kmeans_original = KMeans(n_clusters=n_clusters, random_state=0)
clusters_original = kmeans_original.fit_predict(X_scaled)

# Evaluate clustering performance using silhouette score and Davies–Bouldin index
silhouette_original = silhouette_score(X_scaled, clusters_original)
davies_bouldin_original = davies_bouldin_score(X_scaled, clusters_original)

# Visualize clustering results on original data (using first two features)
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_original, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title(f'K-means Clustering on Original Data (k={n_clusters})')
plt.colorbar(scatter, label='Cluster Label')
plt.show()

# Display the silhouette score and Davies-Bouldin index
silhouette_original, davies_bouldin_original
```

```python
# Apply K-means clustering on the PCA-transformed data (X_pca)
kmeans_pca = KMeans(n_clusters=n_clusters, random_state=0)
clusters_pca = kmeans_pca.fit_predict(X_pca)

# Evaluate clustering performance on PCA data using silhouette score and Davies–Bouldin index
silhouette_pca = silhouette_score(X_pca, clusters_pca)
davies_bouldin_pca = davies_bouldin_score(X_pca, clusters_pca)

# Visualization of clustering results on PCA-transformed data (first two components)
plt.figure(figsize=(10, 7))
scatter_pca = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_pca, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title(f'K-means Clustering on PCA Data (k={n_clusters})')
plt.colorbar(scatter_pca, label='Cluster Label')
plt.show()

# Results: Silhouette and Davies-Bouldin scores for comparison
(silhouette_original, davies_bouldin_original), (silhouette_pca, davies_bouldin_pca)
```

```python
#1. Comparing Clustering Results from Original and PCA Data
#When comparing clustering performance, metrics such as silhouette score and Davies–Bouldin index help to evaluate cluster quality:

#Silhouette Score (higher is better): Measures how similar each point is to its assigned cluster compared to other clusters.
#Davies–Bouldin Index (lower is better): Indicates how well-separated clusters are; a lower value suggests clusters are compact and distinct.
#Typically, PCA can result in a minor loss in silhouette score, but often improves the Davies–Bouldin score due to reduced noise.

#2. Observed Similarities and Differences
#Similarities
#Cluster Structure: Clusters that were distinguishable in the original data often remain so in PCA-transformed data.
#Number of Clusters: Reducing dimensionality should not significantly alter the optimal number of clusters.
#Differences
#Cluster Overlap: After PCA, clusters may be more compact, but the overlap between clusters can increase if dimensionality reduction overly simplifies the data.
#Variance: PCA captures the main variance directions, but some nuanced variations between clusters may be lost.
#3. Impact of Dimensionality Reduction on Clustering Performance
#Reducing dimensions through PCA improves clustering efficiency and sometimes clarity by eliminating less significant features. However, this may cause:

#Loss of Subtle Patterns: Some cluster distinctions may be masked.
#Improved Robustness: Removing noise or weakly contributing features can make clusters more distinct.
#4. Trade-Offs between PCA and Original Data Clustering
#Using PCA before clustering has both benefits and trade-offs:

#Benefits:

#Efficiency: Fewer dimensions lead to faster computations, especially with large datasets.
#Noise Reduction: PCA can filter out features that add little value, improving cluster quality.
#Interpretability: PCA-transformed clusters may show patterns that align better with major data trends.
#Drawbacks:

#Loss of Information: Some data variation may be important for clustering but is lost in PCA.
#Complexity in Interpretation: Clusters in principal component space might not easily map back to original features.
#Generally, PCA is beneficial for large, noisy datasets where a slight loss in detail is acceptable. In datasets with well-defined clusters and minimal noise, clustering on the original standardized data can yield more precise results.
```
