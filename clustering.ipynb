```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(df)
print(normalized_df)
```

```python
normalized_df = pd.DataFrame(normalized_df,columns = df.columns)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(normalized_df)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 4))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Curve for K-Means')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()

# Optimal K (choose based on the elbow point, e.g., K=4 for this example)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(normalized_df)
kmeans_silhouette = silhouette_score(normalized_df, kmeans_labels)
print("K-Means Silhouette Score:", kmeans_silhouette)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate linkage matrix for dendrogram
plt.figure(figsize=(10, 5))
for method in ['ward', 'complete', 'average']:
    linkage_matrix = linkage(normalized_df, method=method)
    dendrogram(linkage_matrix)
    plt.title(f'Hierarchical Clustering Dendrogram ({method} linkage)')
    plt.xlabel("Sample index")
    plt.ylabel("Distance")
    plt.show()

# Choose a linkage and apply AgglomerativeClustering (e.g., 'ward' linkage)
hierarchical = AgglomerativeClustering(n_clusters=4, linkage='ward')
hierarchical_labels = hierarchical.fit_predict(normalized_df)
hierarchical_silhouette = silhouette_score(normalized_df, hierarchical_labels)
print("Hierarchical Clustering Silhouette Score:", hierarchical_silhouette)
```

```python
# DBSCAN with parameter tuning
eps_values = [0.3, 0.5, 0.7]
min_samples_values = [5, 10, 15]

best_silhouette = -1
best_eps = 0
best_min_samples = 0
best_labels = None

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan_labels = dbscan.fit_predict(normalized_df)
        
        # DBSCAN assigns -1 to noise, which can skew silhouette calculation
        if len(set(dbscan_labels)) > 1:  # Ensure more than one cluster exists
            silhouette = silhouette_score(normalized_df, dbscan_labels)
            if silhouette > best_silhouette:
                best_silhouette = silhouette
                best_eps = eps
                best_min_samples = min_samples
                best_labels = dbscan_labels

print("Best DBSCAN Silhouette Score:", best_silhouette)
print("Best DBSCAN parameters: eps =", best_eps, ", min_samples =", best_min_samples)
```

```python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(6, 6))
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="viridis", s=50)
    plt.title(title)
    plt.show()

# Visualize clusters from each algorithm
plot_clusters(normalized_df, kmeans_labels, "K-Means Clustering")
plot_clusters(normalized_df, hierarchical_labels, "Hierarchical Clustering")
plot_clusters(normalized_df, best_labels, "DBSCAN Clustering")
```
